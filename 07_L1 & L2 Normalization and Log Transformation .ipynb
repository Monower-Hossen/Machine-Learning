## üîπ **1. L1 Normalization (Least Absolute Deviations)**

Also known as **Manhattan Norm** or **Least Absolute Shrinkage**, L1 normalization **scales** each row (sample) such that the **sum of absolute values** equals 1.

### üìê **Formula:**

$$
x_{\text{norm}} = \frac{x}{\|x\|_1} = \frac{x}{\sum |x_i|}
$$

### ‚úÖ **Use Cases:**

* When **sparse features** are important.
* In **text classification** (e.g., TF-IDF vectors).






```python
from sklearn.preprocessing import Normalizer
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = [[1, 1, 1], [1, 1, 0], [1, 0, 0]]
l1_normalizer = Normalizer(norm='l1')
print(l1_normalizer.transform(data))
```

## üîπ **2. L2 Normalization (Least Squares)**

Also known as **Euclidean Norm**, L2 normalization scales each row such that the **sum of squared values equals 1**.

### üìê **Formula:**

$$
x_{\text{norm}} = \frac{x}{\|x\|_2} = \frac{x}{\sqrt{\sum x_i^2}}
$$

### ‚úÖ **Use Cases:**

* When **distance-based algorithms** (like KNN, SVM) are used.
* For **gradient descent convergence**.


```python
l2_normalizer = Normalizer(norm='l2')
print(l2_normalizer.transform(data))
```

## üîπ **3. Log Transformation**

**Log Transformation** reduces **skewness** and helps convert **non-normal distributions** to approximately **normal** (bell-shaped).

### üìê **Formula:**

$$
x_{\text{log}} = \log(x + 1)
$$

(We add 1 to avoid log(0))

### ‚úÖ **Use Cases:**

* When data is **right-skewed** (e.g., income, population).
* In regression to **stabilize variance** and improve performance.



```python
import numpy as np
import pandas as pd

data = pd.DataFrame({'Income': [1,5,10,20,50,100,200,500,1000,2000,5000,10000,20000,50000,100000]})
data['Log_Income'] = np.log1p(data['Income'])  # log(x + 1)
print(data)
```

```python
for col in data.columns:
    sns.histplot(data[col], kde=True)
    plt.show()
```
